```
from sklearn import tree
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import numpy as np
from os import listdir
from scipy.misc import *
import imageio
from skimage import transform,io
import tensorflow as tf
from tensorflow import keras
from numpy import array
from sklearn.model_selection import KFold

test_size = 0.2
```

```
def get_img(data_path):
    img = io.imread(data_path, as_gray=True)
    img = transform.resize(img, (100,100), mode='constant', preserve_range=True)
    return img

def print_img(img):
    imgplot = plt.imshow(img, cmap=plt.cm.gray)
    plt.show()
```

```
def get_dataset(dataset_path='./Dataset/'):
    # Getting all data from data path:
    try:
        print("in try")
        X = np.load('npy_dataset/X.npy')
        Y = np.load('npy_dataset/Y.npy')
    except:
        print("inside")
        labels = listdir(dataset_path) # Geting labels
        X = []
        Y = []
        for i, label in enumerate(labels):
            datas_path = dataset_path+'/'+label
            for data in listdir(datas_path):
                img = get_img(datas_path+'/'+data)
                X.append(img)
                Y.append(int(label))
        # Create dateset:
        print_img(X[1])
        print(Y[1])
        for x in range(len(X)):
            X[i] = 1-np.array(X[i]).astype('float64')/255.
            
        Y = np.array(Y).astype('float64')
        
        if not os.path.exists('npy_dataset/'):
            os.makedirs('npy_dataset/')
        np.save('npy_dataset/X.npy', X)
        np.save('npy_dataset/Y.npy', Y)
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=42)
    X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.25, random_state=1)
    return X_train, X_val, X_test, Y_train, Y_val, Y_test
```

```
train_x, val_x, test_x, train_y, val_y, test_y = get_dataset('Sign-Language-Digits-Dataset/Dataset')
```

```
train_x = np.array(train_x)
train_x = np.stack(train_x)
train_x.shape
```

```
print(train_x.shape)
print(train_y.shape)
print(np.asarray(train_y).shape)
```

```
  test_x = np.array(test_x)
  test_x = np.stack(test_x)
  test_x.shape
```

```
def train_test_model(layers, neurons, train_x, val_x, train_y, val_y):
  model = keras.Sequential()
  # input layer
  model.add(keras.layers.Flatten(input_shape=(100, 100)))

  # hidden layers
  for i in range(int(layers)):
    model.add(keras.layers.Dense(neurons, activation='relu'))

  # output layer
  model.add(keras.layers.Dense(10, activation='softmax'))

  model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

  model.fit(train_x, train_y, epochs=10)


  val_test_loss, val_test_acc = model.evaluate(val_x,  val_y, verbose=2)

  print('\nValidation accuracy for model (', layers, ' layers,' , neurons ,' neurons):', val_test_acc * 100.0)
  print('==================================================================================')

  test_loss, test_acc = model.evaluate(test_x,  test_y, verbose=2)
  print('\nTest accuracy for model (', layers, ' layers', '):', test_acc * 100.0)
  print('==================================================================================')

  return test_acc
```

```
avg1 = 0
avg2 = 0
avg3 = 0
avg4 = 0
kfold = KFold(7, True, 1)
for train, test in kfold.split(train_x):
  avg1 += train_test_model(1,128, train_x[train], train_x[test], train_y[train], train_y[test])
```

```
for train, test in kfold.split(train_x):
  avg2 += train_test_model(2,128, train_x[train], train_x[test], train_y[train], train_y[test])
```

```
for train, test in kfold.split(train_x):
  avg3 += train_test_model(1,64, train_x[train], train_x[test], train_y[train], train_y[test])
```

```

for train, test in kfold.split(train_x):
  avg4 += train_test_model(2,64, train_x[train], train_x[test], train_y[train], train_y[test])
```
